```python
X=data.drop(['Attrition'],axis=1)
Y=data['Attrition'].values
```

```python
from collections import Counter
print(Counter(Y))
rus=RandomOverSampler(random_state=42)
X_over, Y_over =rus.fit_resample(X,Y)
print(Counter(Y_over))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test =train_test_split(X_over,Y_over,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve,roc_auc_score
```

```python
X_train_encoded = pd.get_dummies(X_train)
```

```python
logreg=LogisticRegression()
logreg.fit(X_train_encoded,Y_train)
```

```python
X_train_encoded = pd.get_dummies(X_train)
X_test_encoded = pd.get_dummies(X_test)
```

```python
missing_cols = set(X_train_encoded.columns) - set(X_test_encoded.columns)
for col in missing_cols:
    X_test_encoded[col] = 0
```

```python
prediction = logreg.predict(X_test_encoded)
cnf_matrix = confusion_matrix(Y_test, prediction)
print("Accuracy Score:", accuracy_score(Y_test, prediction))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, roc_auc_score

# Plot confusion matrix
fig = plt.figure(figsize=(15, 6))

ax1 = fig.add_subplot(1, 2, 1)
ax1 = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap='Blues', fmt='d')
bottom, top = ax1.get_ylim()
ax1.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel('Predicted')
plt.ylabel('Expected')

# Plot ROC curve
ax2 = fig.add_subplot(1, 2, 2)
y_pred_proba = logreg.predict_proba(X_test_encoded)[:, 1]  # Use encoded test data
fpr, tpr, _ = roc_curve(Y_test, y_pred_proba)
auc = roc_auc_score(Y_test, y_pred_proba)
ax2.plot(fpr, tpr, label="data 1, auc=" + str(auc))
plt.legend(loc=4)
plt.show()
```
